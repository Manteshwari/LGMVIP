#**2) Image to Pencil Sketch with Python :**

##Importing libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

##Reading the Image


```python
image = cv2.imread("unity is strength.jpg")
```

```python
image= cv2.resize(image, (500, 300))
```

```python
cv2_imshow(image)
```

##converting the original image to greyscale:

```python
gray_scale_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray_scale_image)
```

##Inverting the gray scale image

```python
inverted_image = cv2.bitwise_not(gray_scale_image)  # using bitwise_not func to invert the image

cv2_imshow(inverted_image)
```

##Image smoothing

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

##Getting the final pencil sketch

```python
def pencil_sketch(x, y):

    return cv2.divide(x, 255 - y, scale=256)

pencil_sketch=pencil_sketch(gray_scale_image,img_smoothing)
cv2_imshow(pencil_sketch)
```

##And finally, if we want to have a look at both the original image and the pencil sketch image.

```python
cv2_imshow(image)

cv2_imshow(pencil_sketch)
```
